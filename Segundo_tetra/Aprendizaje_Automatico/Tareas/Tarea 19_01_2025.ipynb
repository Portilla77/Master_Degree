#### Problema 1

```python
from typing import Union

def celsius_a_fahrenheit(celsius: Union[int, float]) -> float:
  """Función que convierte grados Celsius y los convierte a Fahrenheit
    Args:
        celsius (int|float): Temperatura en escala Celsius
    Return:
        fahrenheit (int|float): Grados Celsius convertidos a Fahrenheit """
  fahrenheit = celsius*(9/5)+32
  return fahrenheit

grados_celsius = 50

print(f" {grados_celsius}° Celsius convertidos a Fahrenheit son {celsius_a_fahrenheit(grados_celsius)}°")
```

#### Problema 3

```python
from typing import Union
import numpy as np

def calcular_promedio(lista_numeros: list) -> Union[float,None]:
  """Función que toma una lista de valores numéricos y calcula su promedio.
     Omite los valores None de la listaPara el caso de None retornará dicho valor"""

  valores_admisibles= [zero for zero in lista_numeros if zero is not None]
  if not valores_admisibles:
    print("La lista contiene solo valores None")
    return None
  promedio = np.mean(valores_admisibles)
  return f"El promedio es: {promedio}"

valores = [None]
valores = [1,5]
#print(f"El promedio es: {calcular_promedio(valores)}")
calcular_promedio(valores)
```
